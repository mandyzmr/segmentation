#### Data
Train on [Oxford Pets - IIT dataset](https://www.robots.ox.ac.uk/~vgg/data/pets/), including pet images, their classes (37 breeds), segmentation masks (pet, bg, outline) and head region-of-interest. 

```python
(train, test), info = tfds.load('oxford_iiit_pet:3.2.0', with_info=True, split=['train','test'])
print(train)
print(test)
```

```python
#我们不研究37种动物，只研究seg mask的三种label
class_names = ['pet', 'background', 'outline']
n_classes = len(class_names)
n_train = info.splits['train'].num_examples
n_test = info.splits['test'].num_examples
info
```

```python
def preprocess(dataset): #annot mask就是dataset的y，需要同时处理
    images, annots = dataset['image'], dataset['segmentation_mask'] 
    images = tf.image.resize(images, (128, 128), method='nearest') 
    images = tf.reshape(images, [128,128,3]) 
    images = tf.cast(images, tf.float32) 
    images = images/255 #normalize [0,1]
    
    annots = tf.image.resize(annots, (128, 128), method='nearest')
    annots = tf.reshape(annots, [128,128,1])
    annots = tf.cast(annots, tf.int32) #每个像素用label代表，先resize再转dtype
    annots = annots -1 #因为原来的annot label从1开始，统一为0,1,2
    #annots = tf.one_hot(annots[...,-1], depth=n_class, dtype=tf.int32) #转成one-hot （224，224，12)
    return images, annots

def random_flip(images, annots): #增加多样性
    if tf.random.uniform(shape=()) > 0.5: #0-1之间随机实数
        images = tf.image.flip_left_right(images) 
        annots = tf.image.flip_left_right(annots)
    return images, annots
```

```python
AUTO = tf.data.experimental.AUTOTUNE
def prepare_train_dataset(train):
    train = (train
             .map(preprocess, num_parallel_calls=AUTO)
             .map(random_flip, num_parallel_calls=AUTO)
             .shuffle(BUFFER_SIZE) 
             .batch(BATCH_SIZE) 
             .repeat()
             .prefetch(AUTO)) 
    return train

def prepare_test_dataset(test):
    test = (test
            .map(preprocess) 
            .batch(BATCH_SIZE)
            .repeat())
    return test
```

```python
BATCH_SIZE = 64
BUFFER_SIZE = 1000
train = prepare_train_dataset(train)
test = prepare_test_dataset(test)
print(train)
print(test)
```

```python
def plot_segmentation_gray(dataset): #单通道显示
    val = dataset.unbatch().shuffle(BUFFER_SIZE)
    plt.figure(figsize=(8, 4))
    for i, (image, annot) in enumerate(val.take(1)):
        image, annot = image.numpy(), annot.numpy()
        seg_img = annot.reshape(128,128).astype('uint8')
        image = image*255 #denormalize到(0,255)
        image = image.astype('uint8')
        plt.subplot(1,2,1)
        plt.imshow(image)
        plt.axis('off')
        plt.subplot(1,2,2)
        plt.imshow(seg_img) 
        plt.axis('off')
```

#### TF Model
Skip connections concatenate corresponding encoder block outputs to each stage of the decoder. 

<img src='https://drive.google.com/uc?export=view&id=1BeQSKL2Eq6Fw9iRXsN1hgunY-CS2nH7V' alt='unet' style='height:350px;width:550px'>

**Encoder**   
可以考虑把encoder逐步降维hw增加c的过程，用ResNet替代输出中间值

```python
# Same Conv版本
def encoder_block(inputs, block, filters, kernel_size=3, activation='relu', pool_size=2, 
          pool_stride=2, dropout=0.3):
    c = inputs
    for i in range(2):
        c = layers.Conv2D(filters, kernel_size, activation=activation, kernel_initializer='he_normal', 
                          padding='same', name=f'conv{block}_{i+1}')(c)
    p = layers.MaxPooling2D(pool_size, strides=pool_stride, name=f'pooling{block}')(c)
    p = layers.Dropout(dropout, name=f'dropout{block}')(p)
    return c, p


def encoder(inputs):
    c1, p1 = encoder_block(inputs, '1', filters=64) #(128,128)->(64,64)
    c2, p2 = encoder_block(p1, '2', filters=128) #(64,64)->(32,32)
    c3, p3 = encoder_block(p2, '3', filters=256) #(32,32)->(16,16)
    c4, p4 = encoder_block(p3, '4', filters=512) #(16,16)->(8,8)
    return p4, (c1, c2, c3, c4)
```

**Bottleneck**

```python
def bottleneck(inputs, block, filters, kernel_size=3, activation='relu'):
    x = inputs
    for i in range(2):
        x = layers.Conv2D(filters, kernel_size, activation=activation, kernel_initializer='he_normal', 
                          padding='same', name=f'conv{block}_{i+1}')(x)
    return x
```

**Decoder**

```python
def decoder_block(inputs, conv_output, block, filters, kernel_size=4, strides=2, dropout=0.3):
    #tf的padding神奇在于，当padding='same'且strides!=1时，仍然可以翻倍而非wh不变
    #可以用transpose或者 Upsampling(bilinear, 2) + 1x1 conv2d
    x = layers.Conv2DTranspose(filters, kernel_size, strides, padding='same', 
                               name=f'conv_transpose{block}')(inputs) #(8,8)->(18,18)
    x = layers.Cropping2D(cropping=(1,1))(x) #(18,18)->(16,16)
    
    #得到和x1一样的channel
    x = layers.concatenate([x, conv_output], name=f'concat{block}')
    x = layers.Dropout(dropout, name=f'dropout{block}')(x)
    for i in range(2):
        x = layers.Conv2D(filters, kernel_size, activation='relu', kernel_initializer='he_normal', 
                          padding='same', name=f'conv{block}_{i+1}')(x)
    return x
    

def decoder(inputs, conv_outputs, n_classes):
    (c1, c2, c3, c4) = conv_outputs
    c6 = decoder_block(inputs, c4, '6', filters=512)
    c7 = decoder_block(c6, c3, '7', filters=256)
    c8 = decoder_block(c7, c2, '8', filters=128)
    c9 = decoder_block(c8, c1, '9', filters=64)
    x = layers.Conv2D(n_classes, 1, activation='softmax', name=f'conv10')(c9)
    return x
```

```python
def Unet():
    inputs = keras.Input(shape=(128,128,3))
    encoder_output, convs = encoder(inputs)
    bottle_neck = bottleneck(encoder_output, '5', filters=1024)
    outputs = decoder(bottle_neck, convs, n_classes)
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
model = Unet()
model.summary()
```

```python
STEPS_PER_EPOCH = n_train//BATCH_SIZE
VALIDATION_STEPS = n_test//BATCH_SIZE
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(train, steps_per_epoch=STEPS_PER_EPOCH, epochs=10,
                    validation_data=test, validation_steps=VALIDATION_STEPS)
```

#### PyTorch Model

```python
class ConvBlock(nn.Module):
    def __init__(self, in_chans, out_chans, padding):
        super(ConvBlock, self).__init__()
        block = []
        for i in range(2):
            block+=[
                nn.Conv2d(in_chans, out_chans, kernel_size=3, padding=int(padding)),
                nn.BatchNorm2d(out_chans), 
                nn.ReLU(inplace=True)
            ]
            in_chans = out_chans
        self.block = nn.Sequential(*block) #把元素都拿出来，并非列表

    def forward(self, x):
        x = self.block(x)
        return x
```

```python
class UpBlock(nn.Module):
    def __init__(self, in_chans, out_chans, up_mode, padding):
        super(UpBlock, self).__init__()
        if up_mode == 'upconv': #转置
            self.up = nn.ConvTranspose2d(in_chans, out_chans, kernel_size=2, stride=2)
        elif up_mode == 'upsample': #双线性插入
            self.up = nn.Sequential(
            nn.Upsample(mode='bilinear', scale_factor=2),
            nn.Conv2d(in_chans, out_chans, kernel_size=1)) #运算后再融合一下features
        
        self.conv_block = ConvBlock(in_chans, out_chans, padding)
      
    def forward(self, x, bridge):
        up = self.up(x)
        crop = transforms.CenterCrop((up.size(2),up.size(3)))(bridge) 
        out = torch.cat([crop, up], dim=1)
        out = self.conv_block(out)
        return out
```

```python
class UNet(nn.Module):
    def __init__(self, in_chans=3, n_classes=3, padding=False, up_mode='upconv'):
        super(UNet, self).__init__()
        self.n_classes = n_classes
        self.padding = padding #做same/valid conv
        self.up_mode = 'upconv' #上采样方式
        assert self.up_mode in ('upconv', 'upsample')

        # Encoder
        self.encoder = nn.ModuleList() #用循环替代逐个写出
        out_chans = 64
        for i in range(5): #16倍下采样，最后一层不加pool
            self.encoder.append(ConvBlock(in_chans, out_chans, self.padding))
            in_chans = out_chans
            out_chans *= 2 #(128, 256, 512, 1024)
        
        # Decoder
        self.decoder = nn.ModuleList()
        for i in range(4): #16倍上采样
            self.decoder.append(UpBlock(in_chans, in_chans//2, self.up_mode, self.padding))
            in_chans //= 2 #(512, 256, 128, 64)

        self.cls_conv = nn.Conv2d(in_chans, self.n_classes, kernel_size=1)
        
        # 初始化参数
        for m in self.modules():
            if isinstance(m, nn.Conv2d): #针对Conv relu
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.BatchNorm2d): #针对BatchNorm
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        # Encoder
        bridges = []
        for i, encode_layer in enumerate(self.encoder):
            x = encode_layer(x)
            if i < len(self.encoder) - 1: #前4个block
                bridges.append(x) #把pool前的结果记录下来
                x = nn.MaxPool2d(kernel_size=2, stride=2)(x) #pool层没有参数，可以放到这里

        # Decoder
        for i, decode_layer in enumerate(self.decoder):
            x = decode_layer(x, bridges[-i-1]) #逆序取出
        
        x = self.cls_conv(x)
        score = nn.Softmax(dim=1)(x)
        
        return score
```

```python
model = UNet(padding=True)
summary(model, (3, 128, 128))
```

#### ResNet UNet
<a id='resnet_block'></a>

```python
# 适用于ResNet18,34
class BasicBlock(nn.Module):
    expansion = 1 #为了和Bottleneck保持一致，方便引用
    def __init__(self, in_chans, out_chans, stride=1, kernel_size=3, dilation_rate=1): 
        super(BasicBlock, self).__init__()
        #第一层stride有时候需要downsample，而第二层stride默认1
        self.block = nn.Sequential(
                nn.Conv2d(in_chans, out_chans, kernel_size, stride, padding=1*dilation_rate, 
                          dilation=dilation_rate, bias=False),
                nn.BatchNorm2d(out_chans), 
                nn.ReLU(inplace=True),
                nn.Conv2d(out_chans, out_chans*BasicBlock.expansion, kernel_size, padding=1, bias=False),
                nn.BatchNorm2d(out_chans*BasicBlock.expansion))
        
        #当downsampling时，需要用ConvBlock
        if stride != 1 or in_chans != BasicBlock.expansion * out_chans:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_chans, out_chans*BasicBlock.expansion, 1, stride, bias=False),
                nn.BatchNorm2d(out_chans*BasicBlock.expansion))
        else:
            self.shortcut = nn.Sequential()
        
    def forward(self, inputs):
        x = self.block(inputs)
        x = nn.ReLU(inplace=True)(x + self.shortcut(inputs))
        return x
```

```python
# 适用于ResNet50,101,152
class BottleNeck(nn.Module):
    expansion = 4 
    def __init__(self, in_chans, out_chans, stride=1, kernel_size=3, dilation_rate=1): 
        super(BottleNeck, self).__init__()
        #瓶颈 feature map
        self.block =  nn.Sequential(
                nn.Conv2d(in_chans, out_chans, kernel_size=1, bias=False),
                nn.BatchNorm2d(out_chans), 
                nn.ReLU(inplace=True),
                nn.Conv2d(out_chans, out_chans, kernel_size, stride, padding=1*dilation_rate, 
                          dilation=dilation_rate, bias=False),
                nn.BatchNorm2d(out_chans), 
                nn.ReLU(inplace=True),
                nn.Conv2d(out_chans, out_chans*BottleNeck.expansion, kernel_size=1, bias=False),
                nn.BatchNorm2d(out_chans*BottleNeck.expansion))
       
        #当downsampling时，需要用ConvBlock
        if stride != 1 or in_chans != BottleNeck.expansion * out_chans:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_chans, out_chans*BottleNeck.expansion, 1, stride, bias=False),
                nn.BatchNorm2d(out_chans*BottleNeck.expansion))
        else:
            self.shortcut = nn.Sequential()
            
    def forward(self, inputs):
        x = self.block(inputs)
        x = nn.ReLU(inplace=True)(x + self.shortcut(inputs))
        return x
```

```python
class ResNet(nn.Module): #没有FC，16倍下采样，用作UNet backbone
    def __init__(self, name, in_chans=3, block=None, n_blocks=None): 
        super(ResNet, self).__init__() 
       
        # 结构：使用自定义的结构，或者经典结构
        self.structure = name
        self.block = block #用basicblock还是bottleneck
        self.n_blocks = n_blocks #每个block循环次数
        self.get_structure() #重新整理结构
        strides=[1,2,2,2] 
        
        # Stem：由于UNet是16倍下采样，把ResNet原有的7x7Conv改为3x3Conv
        self.layer1 = nn.Sequential(
             nn.Conv2d(in_chans, 64, kernel_size=3, padding=1, bias=False),
                nn.BatchNorm2d(64), 
                nn.ReLU(inplace=True))
        self.pool = nn.MaxPool2d(3, 2, padding=1) #2倍downsample
        
        #除了layer2因为有事先pooling降维之外，其余layer都是第一个block的stride=2降维，其余blocks的stride=1
        self.in_chans = 64 #后续不断更新
        self.layer2 = self.build_blocks(64, self.n_blocks[0], stride=strides[0])         
        self.layer3 = self.build_blocks(128, self.n_blocks[1], stride=strides[1]) #4倍downsample  
        self.layer4 = self.build_blocks(256, self.n_blocks[2], stride=strides[2]) #8倍downsample  
        self.layer5 = self.build_blocks(512, self.n_blocks[3], stride=strides[3]) #16倍downsample
    
    def get_structure(self):
        versions = {'resnet18':(BasicBlock,[2,2,2,2]), 
                     'resnet34':(BasicBlock,[3,4,6,3]),
                     'resnet50':(BottleNeck,[3,4,6,3]),
                     'resnet101':(BottleNeck,[3,4,23,3]),
                     'resnet152':(BottleNeck,[3,8,36,3])}
        
        if not self.block or not self.n_blocks: #如果任意一项没有定义
            self.block, self.n_blocks = versions[self.structure]
        
        
    def build_blocks(self, out_chans, n_blocks, stride):      
        strides = [stride] + [1]*(n_blocks-1) #除了第一个block以外，其余stride=1
        blocks = []
        for i in range(n_blocks):           
            blocks.append(self.block(self.in_chans, out_chans, stride=strides[i])) 
            self.in_chans = out_chans*self.block.expansion
        blocks = nn.Sequential(*blocks) 
        return blocks
 

    def forward(self, inputs): 
        f1 = self.layer1(inputs) #224
        f2 = self.layer2(self.pool(f1)) #2
        f3 = self.layer3(f2) #4        
        f4 = self.layer4(f3) #8        
        f5 = self.layer5(f4) #16
        return f1, f2, f3, f4, f5 #返回4个bridges和一个bottleneck
    
```

```python
model = ResNet('resnet152')
summary(model, (3,224,224))
```

```python
class UpBlock(nn.Module):
    def __init__(self, in_chans, out_chans, bridge_chans, up_mode):
        super(UpBlock, self).__init__()
        if up_mode == 'upconv': #转置
            self.up = nn.ConvTranspose2d(in_chans, out_chans, kernel_size=2, stride=2)
        elif up_mode == 'upsample': #双线性插入
            self.up = nn.Sequential(
            nn.Upsample(mode='bilinear', scale_factor=2),
            nn.Conv2d(in_chans, out_chans, kernel_size=1)) #运算后再融合一下features
        
        #用BasicBlock代替之前的ConvBlock，不需要定义padding，out_chans和Brige_chans不一定相等，需要分开指定
        self.conv_block = BasicBlock(out_chans+bridge_chans, out_chans) 
      
    def forward(self, x, bridge):
        up = self.up(x)
        crop = transforms.CenterCrop((up.size(2),up.size(3)))(bridge) 
        out = torch.cat([crop, up], dim=1)
        out = self.conv_block(out)
        return out
```

```python
class ResNet_UNet(nn.Module):
    def __init__(self, in_chans=3, n_classes=3, up_mode='upconv', backbone='resnet50'):
        super(ResNet_UNet, self).__init__()
        self.n_classes = n_classes
        self.up_mode = 'upconv' #上采样方式
        assert self.up_mode in ('upconv', 'upsample')

        # Encoder
        self.encoder = ResNet(backbone, in_chans) #16倍下采样 (64,)+ (64, 128, 256, 512)*4
        in_chans = 512 * self.encoder.block.expansion #ResNet每层的输出channel
        
        # Decoder
        self.decoder = nn.ModuleList()
        for i in range(3): #16倍上采样
            self.decoder.append(UpBlock(in_chans, in_chans//2, in_chans//2, self.up_mode))
            in_chans //= 2 #(256, 128, 64)*4
        #最后一步in&out都是64，不是in_chans//2，所以UpBlock里拼接之后的in_chans和原来in_chans不一定相等
        self.decoder.append(UpBlock(in_chans, 64, 64, self.up_mode)) 
        
        self.cls_conv = nn.Conv2d(64, self.n_classes, kernel_size=1)
        
        # 初始化参数
        for m in self.modules():
            if isinstance(m, nn.Conv2d): #针对Conv relu
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.BatchNorm2d): #针对BatchNorm
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        # Encoder
        f1, f2, f3, f4, f5 = self.encoder(x)
        bridges = [f1, f2, f3, f4]
        x = f5
        
        # Decoder
        for i, decode_layer in enumerate(self.decoder):
            x = decode_layer(x, bridges[-i-1]) #逆序取出
        
        x = self.cls_conv(x)
        score = nn.Softmax(dim=1)(x)
        
        return score
```

```python
model = ResNet_UNet()
summary(model, (3,224,224))
```

#### ResNet UNet++

<img src='images/unetpp.png'></img>

```python
class UpBlock(nn.Module):
    def __init__(self, in_chans, out_chans, n_bridges, up_mode):
        super(UpBlock, self).__init__()
        if up_mode == 'upconv': #转置
            self.up = nn.ConvTranspose2d(in_chans, out_chans, kernel_size=2, stride=2) #默认转置
        elif up_mode == 'upsample': #双线性插入
            self.up = nn.Sequential(
            nn.Upsample(mode='bilinear', scale_factor=2),
            nn.Conv2d(in_chans, out_chans, kernel_size=1)) #运算后再融合一下features
        
        #用BasicBlock代替ConvBlock，由于有多个bridges通过skip conn加入到Decoder，bridge_chans为多个bridges chans之和
        self.conv_block = BasicBlock(out_chans+(n_bridges*out_chans), out_chans) 
      
    def forward(self, x, bridge_list):
        up = self.up(x)
        out = torch.cat([up]+bridge_list, dim=1)
        out = self.conv_block(out)
        return out
```

```python
class ResNet_UNetpp(nn.Module): 
    def __init__(self, in_chans=3, n_classes=3, up_mode='upconv', backbone='resnet50', l=4):
        super(ResNet_UNetpp, self).__init__()
        self.n_classes = n_classes
        self.up_mode = 'upconv' #上采样方式
        self.l = l
        assert self.up_mode in ('upconv', 'upsample')
        
        # Encoder
        self.encoder = ResNet(backbone, in_chans) #16倍下采样 (64,)+ (64, 128, 256, 512)*4
        fea_chans = [64] + [i*self.encoder.block.expansion for i in [64, 128, 256, 512]] #ResNet每层的输出channel
        
        # Decoder
        if self.l>=1:
            self.x01 = UpBlock(fea_chans[1], fea_chans[0], 1, self.up_mode)
            self.cls_x01 = nn.Conv2d(fea_chans[0], self.n_classes, kernel_size=1)
        if self.l>=2:
            self.x11 = UpBlock(fea_chans[2], fea_chans[1], 1, self.up_mode)
            self.x02 = UpBlock(fea_chans[1], fea_chans[0], 2, self.up_mode)
            self.cls_x02 = nn.Conv2d(fea_chans[0], self.n_classes, kernel_size=1)
        if self.l>=3:
            self.x21 = UpBlock(fea_chans[3], fea_chans[2], 1, self.up_mode)
            self.x12 = UpBlock(fea_chans[2], fea_chans[1], 2, self.up_mode)
            self.x03 = UpBlock(fea_chans[1], fea_chans[0], 3, self.up_mode)
            self.cls_x03 = nn.Conv2d(fea_chans[0], self.n_classes, kernel_size=1)
        if self.l>=4:
            self.x31 = UpBlock(fea_chans[4], fea_chans[3], 1, self.up_mode)
            self.x22 = UpBlock(fea_chans[3], fea_chans[2], 2, self.up_mode)
            self.x13 = UpBlock(fea_chans[2], fea_chans[1], 3, self.up_mode)
            self.x04 = UpBlock(fea_chans[1], fea_chans[0], 4, self.up_mode)
            self.cls_x04 = nn.Conv2d(fea_chans[0], self.n_classes, kernel_size=1)
           
        # 初始化参数
        for m in self.modules():
            if isinstance(m, nn.Conv2d): #针对Conv relu
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.BatchNorm2d): #针对BatchNorm
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        # Encoder
        x00, x10, x20, x30, x40 = self.encoder(x)
        
        # Decoder
        scores = []
        if self.l>=1:
            x01 = self.x01(x10,[x00]) 
            out_x01 = self.cls_x01(x01) 
            scores.append(nn.Softmax(dim=1)(out_x01))
        if self.l>=2:
            x11 = self.x11(x20,[x10]) 
            x02 = self.x02(x11,[x00, x01]) 
            out_x02 = self.cls_x02(x02)
            scores.append(nn.Softmax(dim=1)(out_x02))
        if self.l>=3:
            x21 = self.x21(x30,[x20]) 
            x12 = self.x12(x21,[x10, x11]) 
            x03 = self.x03(x12,[x00, x01, x02]) 
            out_x03 = self.cls_x03(x03)
            scores.append(nn.Softmax(dim=1)(out_x03))
        if self.l>=4:
            x31 = self.x31(x40,[x30]) 
            x22 = self.x22(x31,[x20, x21]) 
            x13 = self.x13(x22,[x10, x11, x12]) 
            x04 = self.x04(x13,[x00, x01, x02, x03]) 
            out_x04 = self.cls_x04(x04)
            scores.append(nn.Softmax(dim=1)(out_x04))
        return scores
    
```

```python
def plot_prediction(dataset, n):
    val = dataset.unbatch()
    for (images, annots) in val.take(1):
        y_true = tf.reshape(annots,(-1,128,128))
        y_pred = model.predict(images[n][tf.newaxis,...]) #为减少运算量，只预测一张图(1,128,128,3)
        y_pred = np.argmax(y_pred, axis=-1) #(1,128,128)
        image, y_true, y_pred = images[n].numpy(), y_true[n].numpy(), y_pred[0]
        
    # 对比图片
    image = image*255 #denormalize到(0,255)
    image = image.astype('uint8')
    images = [image, y_true, y_pred]
    titles = ['Image','True Mask','Predicted Mask']
    plt.figure(figsize=(15, 4))
    for i, image in enumerate(images):
        plt.subplot(1,3,i+1)
        plt.imshow(image) 
        plt.title(titles[i])
        plt.axis('off')
    
    # 这张图片的metric
    class_iou, class_dice_score = iou_and_dice_score(y_true, y_pred)
    metrics = pd.DataFrame({'Class':class_names,'IOU':class_iou,'Dice Score':class_dice_score})
    metrics = metrics.sort_values(['IOU','Dice Score'], ascending=False)
    print(metrics)
```
